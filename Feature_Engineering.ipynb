

**1. What is a parameter?**

* A parameter is a numerical value that defines a model’s behavior and is learned from training data in machine learning (e.g., weights in linear regression).

**2. What is correlation?**

* Correlation measures the relationship between two variables, indicating how one variable moves in relation to another.

**3. What does negative correlation mean?**

* Negative correlation means that as one variable increases, the other decreases (e.g., more exercise, lower weight).

**4. Define Machine Learning. What are the main components in Machine Learning?**

* Machine Learning is the process of teaching machines to learn patterns from data and make decisions. Components include data, models, loss functions, optimizers, and evaluation metrics.

**5. How does loss value help in determining whether the model is good or not?**

* A lower loss value indicates the model's predictions are closer to the actual values, hence better performance.

**6. What are continuous and categorical variables?**

* Continuous variables have infinite numeric values (e.g., temperature). Categorical variables represent categories or labels (e.g., color, gender).

**7. How do we handle categorical variables in Machine Learning? What are the common techniques?**

* Categorical variables are handled using encoding techniques like Label Encoding, One-Hot Encoding, and Ordinal Encoding.

**8. What do you mean by training and testing a dataset?**

* Training data is used to build the model; testing data evaluates the model’s performance on unseen examples.

**9. What is sklearn.preprocessing?**

* It is a module in scikit-learn that provides functions for feature scaling, encoding, and transforming input data.

**10. What is a Test set?**

* A test set is a portion of the dataset used to assess how well a trained model generalizes to new, unseen data.

**11. How do we split data for model fitting (training and testing) in Python?**

* We use `train_test_split()` from `sklearn.model_selection` to split data into training and testing sets.

**12. How do you approach a Machine Learning problem?**

* Define the problem, gather and clean data, perform EDA, choose a model, train it, evaluate performance, and refine.

**13. Why do we have to perform EDA before fitting a model to the data?**

* Exploratory Data Analysis (EDA) helps understand the data’s structure, detect patterns, and clean errors before model training.

**14. What is correlation?**

* Correlation is a measure of the strength and direction of a linear relationship between two variables.

**15. What does negative correlation mean?**

* It means that as one variable increases, the other tends to decrease.

**16. How can you find correlation between variables in Python?**

* Use `df.corr()` in pandas or `np.corrcoef()` in NumPy to compute the correlation matrix.

**17. What is causation? Explain difference between correlation and causation with an example.**

* Causation implies one event causes another; correlation means they are related. For example, ice cream sales and drowning are correlated but not causally related.

**18. What is an Optimizer? What are different types of optimizers? Explain each with an example.**

* Optimizers adjust model parameters to minimize loss. Common types:

  * **SGD**: Updates weights using gradients.
  * **Adam**: Adaptive learning rate optimizer.
  * **RMSprop**: Uses moving averages of squared gradients.

**19. What is sklearn.linear\_model?**

* It is a module in scikit-learn that provides linear models like LinearRegression, LogisticRegression, etc.

**20. What does model.fit() do? What arguments must be given?**

* `fit()` trains the model on training data. Required arguments: feature matrix `X` and target `y`.

**21. What does model.predict() do? What arguments must be given?**

* `predict()` uses the trained model to make predictions. It requires input features `X`.

**22. What are continuous and categorical variables?**

* Continuous: numeric values with a range (e.g., age). Categorical: distinct categories (e.g., country).

**23. What is feature scaling? How does it help in Machine Learning?**

* Feature scaling standardizes input features so they contribute equally, improving model performance.

**24. How do we perform scaling in Python?**

* Use `StandardScaler` or `MinMaxScaler` from `sklearn.preprocessing`.

**25. What is sklearn.preprocessing?**

* It is a module with tools for scaling, transforming, and encoding data for machine learning.

**26. How do we split data for model fitting (training and testing) in Python?**

**27. Explain data encoding?**
